## To visualize an image in the dataset:

## To visualize multi images in the dataset:

```python
import pickle
import torch
import torchvision
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

with open("../release/train.pkl", "rb") as f:
    data = pickle.load(f) # a dictionary

plt.figure(figsize=(15,10))
start_img = 10000
figure_num = 20
columns = 5
transform = torchvision.transforms.ToPILImage()

for i in range(figure_num):
    plt.subplot(int(figure_num / columns) + 1, columns, i + 1)
    image = transform(torch.tensor(data["images"][start_img+i])*255)
    plt.imshow(image)
    
print("labels:", data["labels"][start_img:start_img+figure_num])
```
